```python

# Análise de Performance Logística com Python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
df = pd.read_csv("base_analise_logistica.csv")

# Visão geral dos dados
print("Resumo do dataset:")
print(df.head())

# Estatísticas básicas
print("\nEstatísticas descritivas:")
print(df.describe())

# Quantidade de entregas por região
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Regiao", palette="Set2")
plt.title("Entregas por Região")
plt.xlabel("Região")
plt.ylabel("Quantidade de Entregas")
plt.tight_layout()
plt.show()

# Status das entregas
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Status", palette="pastel")
plt.title("Status das Entregas")
plt.xlabel("Status")
plt.ylabel("Frequência")
plt.tight_layout()
plt.show()

# Tempo médio de entrega por operador logístico
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Operador_Logistico", y="Tempo_Entrega_dias", estimator="mean", ci=None, palette="Blues_d")
plt.title("Tempo Médio de Entrega por Operador Logístico")
plt.ylabel("Tempo Médio (dias)")
plt.xlabel("Operador Logístico")
plt.tight_layout()
plt.show()
```
